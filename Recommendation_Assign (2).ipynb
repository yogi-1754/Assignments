#Outlier Detection: User defined Function to calculate Upper Extreme and Lower Extreme value

```python
def recommen_fun(data,colname):
    q1 = data[colname].quantile(0.25)
    q3 = data[colname].quantile(0.75)
    iqr = q3-q1

    upper_extreme = q3+(1.5*iqr)
    lower_extreme = q1-(1.5*iqr)

    return lower_extreme, upper_extreme, q1, q3
```

#Evaluate the recommendation system using appropriate metrics such as precision, recall, and F1-score.

```python
from sklearn.metrics import precision_score, recall_score, f1_score


true_labels = [1, 0, 1, 0, 1]  # Whether an anime is relevant (1) or not (0)
predictions = [1, 0, 1, 0, 0]  # Predicted relevance of recommendations

# Calculate Precision, Recall, and F1-score
precision = precision_score(true_labels, predictions)
recall = recall_score(true_labels, predictions)
f1 = f1_score(true_labels, predictions)

print(f'Precision: {precision}, Recall: {recall}, F1-Score: {f1}')
```

## Interview Questions:

#1. Can you explain the difference between user-based and item-based collaborative filtering?

#User-Based Filtering: Recommends items to a user based on what similar users have liked. It computes similarity between users.
Item-Based Filtering: Recommends items similar to those that the user has liked. It computes similarity between items.

#2. What is collaborative filtering, and how does it work?

#Collaborative Filtering: Uses the behavior of multiple users to make recommendations. It works by either finding users similar to the target user or finding items similar to the ones the user liked, then recommending items that those users or items are connected to.
