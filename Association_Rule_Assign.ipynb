## USING TRANSACTION ENCODER:

```python
# Split items into lists by commas
df['items'] = df.iloc[:, 0].str.split(',')
```

```python
df.head()
```

```python
from mlxtend.preprocessing import TransactionEncoder

# Convert the list of items into the format required by the TransactionEncoder
transactions = df['items'].tolist()

# Initialize the TransactionEncoder and fit_transform the data
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)# te.fit_transform(transaction)

# Convert the binary matrix to a DataFrame
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

# Show the first few rows of the encoded data
df_encoded
```

## APRIORI ALGORITHM :

```python
# GENERATE FREQUENT ITEMS USING APRIORI ALGORITHM..

from mlxtend.frequent_patterns import apriori
freq_items=apriori(df_encoded,min_support=0.01,use_colnames=True)
freq_items
```

## GENERATING ASSOCIATION RULES

```python
# GENERATE ASSOCIATION RULES :
from mlxtend.frequent_patterns import association_rules
asso_rules=association_rules(freq_items,metric="confidence",min_threshold=0.5)
asso_rules=asso_rules.sort_values("confidence",ascending=False)
asso_rules

# MIN THRESHOLD IS A HYPER PARAMETER..


# HERE ,
# LIFT RATION SHOULD BE MORE THAN 1
# Rule: (ground beef, eggs) -> (mineral water)

#Support: 0.020000 (2% of the total transactions contain both "ground beef" and "eggs" and also "mineral water").

#Confidence: 0.506667 (In transactions where "ground beef" and "eggs" appear together, there is a 50.67% chance that "mineral water" is also purchased).

#Lift: 2.126469 (The likelihood of purchasing "mineral water" increases by 2.13 times when "ground beef" and "eggs" are bought together, 
#compared to buying "mineral water" randomly).

#Leverage: 0.010133 (This shows the difference in the probability of buying
#"mineral water" when "ground beef" and "eggs" are purchased together versus randomly).

#Conviction: 1.544054 (This means the rule is 1.54 times as likely to make an accurate prediction as a random guess).
```

#NOW we can choose any rule based on your goals:

##High Lift: For strong associations between items.
High Confidence: For reliable rules with a high chance of the consequent item being bought.
High Support: For rules that apply to a large portion of transactions.
High Conviction: For rules with strong predictive power.
Pick the rule that best matches your specific needs, such as cross-selling or product recommendations.

## HYPER PARAMETER TUNING :

```python
# DEFINE RANGES FOR MIN_SUPPORT AND MIN_THRESHOLD..
support_values = [0.01, 0.05, 0.1, 0.2,]
confidence_values = [0.5, 0.6, 0.7, 0.8, 0.9]



# GRID SEARCH ALGORITHMS :

for support in support_values:
    frequent_itemsets = apriori(df_encoded, min_support=support, use_colnames=True)
    for confidence in confidence_values:
        rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=confidence)
        # Evaluate and store the best rules based on lift, confidence, etc.
        print(f"Support: {support}, Confidence: {confidence}, Rules: {len(rules)}")
```

```python
# For min_support >= 0.05, no rules are generated at any confidence level.
# FINAL WE SELECT MINIMUM SUPPORT 0.01 AND CONFIDENCE 0.5..
# OR WE  can use lower values for both min_support and min_confidence to generate more rules.
```

## WHAT IS LIFT AND WHY IT IS IMPORTANT IN ASSOCIATION RULES ?

#ift is a metric used in association rule mining to evaluate the strength of a rule by comparing the observed co-occurrence of items to what would be expected if they were independent. It tells you how much more likely two items are to be purchased together compared to if their co-occurrence were purely random.
Confidence 
A→B: The probability of buying item B given that item A is bought.

Support 
B: The proportion of transactions that contain item B.

## WHAT IS SUPPORT AND CONFIDENCE AND HOW DO I CALCULATE THEM

#Support and Confidence are key metrics in association rule mining that help evaluate the strength and relevance of rules generated from a dataset.

#1. Support

#Support measures how frequently an itemset appears in the dataset. It helps identify the general importance or frequency of an itemset, or combination of items, in all transactions.

#2. Confidence

#Confidence measures the likelihood that item B is purchased when item A is already purchased. It reflects the conditional probability of B given A.

#We can calculate Support and Confidence using the Apriori algorithm or other tools such as mlxtend for association rule mining.
