# PPS (PREDICTIVE POWER SCORE)

```python
## THE MAIN DIFFERENCE BETWEEN CORRELATION AND PREDICTIVE POWER SCORE LIES IN WHAT THEY MEASURE AND HOW THEY ARE USED
```

```python
### CORRELATION :
- **CORRELATION MEASURES THE LINEAR RELATIONSHIP BETWEEN TWO NUMERICAL VARIABLES**

- **IT RANGES FROM -1 TO 1**

- **VALUE CLOSE TO 1 INDICATE A STRONGE POSITIVE LINEAR RELATIONSHIP**

- **VALUE CLOSE TO -1 INDICATES A STRONGE NEGATIVE LINEAR RELATIONSHIP**

- **VALUE CLOSE TO 0 MEANS NO LINEAR RELATIONSHIP**

- **CORRELATION IS SYMMETRIC,MEANS THE RELATIONSHIP BETWEEN VARIABLE IS THE SAME BOTH WAYS,BEACUSE IT MEASURE THE LINEAR RELATIONSHIP BETWEEN TWO VARIABLE WITHOUT CONSIDERING DIRECTION**

- **HOW THIS METHOD IS WORKING ALSO MATERS: AS CORRELATION CALCULATE THE STRENGTH AND DIRECTION OF LINEAR RELATIONSHIP USING A FORMULA THAT IS INHERENTLY SYMMETRIC**
```

```python
# CALCULATE PPSCORE:
# CALCULATING WHOLE PPS MATRIX :
import ppscore as pps
pps.matrix(data)

# TO CONVERT PPS MATRIX INTO DATAFRAME
DATAFRAME=pd.DataFrame(pps.matrix(data))

```

```python
DATAFRAME
```

```python
# TO FILTER OUT SELF PREDICTING CASES(e.g.:AGE TO AGE)
FILTERED_DATAFRAME=DATAFRAME[DATAFRAME.case != "predict_itself"]
FILTERED_DATAFRAME
```

```python
# SELECTING ONLY REQUIRED COLUMNS :
FILTERED_DATAFRAME=FILTERED_DATAFRAME[["x","y","ppscore","case","metric","model"]]
FILTERED_DATAFRAME
```

```python
# SORTING THE DATA BY PPSCORE:
PPSCORE_DATA=FILTERED_DATAFRAME.sort_values("ppscore",ascending=False)
PPSCORE_DATA
```
