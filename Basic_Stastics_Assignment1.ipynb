## mean, median, mode, and standard deviation for numerical column

```python
# Calculate mean, median, mode, and standard deviation
M = df[num_col].mean()
Me = df[num_col].median()
Mode = df[num_col].mode().iloc[0]  
std_dev = df[num_col].std()


print("Mean:\n", M)
print("\nMedian:\n", Me)
print("\nMode:\n", Mode)
print("\nStandard Deviation:\n", std_dev)
```

##  Interpretation of Statistics

```python
#Mean() = This is the sum of all data points deivided by the number of points.
It gives a central value of the dataset.

#Median() = The middle value when the data points are arranged in ascending or descending order.

#Mode() = The most frequently occuring value in the dataset.

#Range = The difference between the highest and lowest values. A larger range indicates more variability in the data.

#Standard Deviation = Measures the spread of data points around the mean. 
    A high standard deviation means data points are spread out widely,
    while a low standard deviation means they are close to the mean .

#Variance = The square of the standard deviation. It also measures the spread of data points.
```
